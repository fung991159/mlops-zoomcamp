## Q2. Computing duration

Now let's compute the `duration` variable. It should contain the duration of a ride in minutes. 

What's the standard deviation of the trips duration in January?

* 32.59
* 42.59
* 52.59
* 62.59

```python
df_train.loc[:, 'duration'] = df_train.tpep_dropoff_datetime - df_train.tpep_pickup_datetime
df_train.loc[:, 'duration_in_min'] = df_train.duration.apply(lambda td: td.total_seconds() / 60)
```

```python
df_train.duration_in_min.std()
# standard deviation of the trips duration in January is 42.59 minute
```

## Q3. Dropping outliers
